```python
import sqlite3
```

```python
con = sqlite3.connect("RecipeDB.db")
cur = con.cursor()
```

```python

sql1 = """
    select recID from Recipe where recDisc = ""
    or recSteps = 0
    or recLink is NULL
    or recName like "%test%"
"""
sqlDelAlleInRec = """
    Delete from AlleInRec where recID = ?;
"""
sqlDelIngInRec = """
    Delete from IngInRec where recID = ?;
"""
sqlDelNutrInRec = """
    Delete from NutrInRec where recID = ?;
"""
sqlDelStepsInRec = """
    Delete from StepsInRec where recID = ?;
"""
sqlDelTagsInRec = """
    Delete from TagsInRec where recID = ?;
"""
for id in cur.execute(sql1).fetchall():
    cur.execute(sqlDelAlleInRec, id)
    cur.execute(sqlDelIngInRec, id)
    cur.execute(sqlDelNutrInRec, id)
    cur.execute(sqlDelStepsInRec, id)
    cur.execute(sqlDelTagsInRec, id)
    print(id)
    

sql1 = """
    Delete from Recipe where recDisc = ""
    or recSteps = 0
    or recLink is NULL
    or recName like "%test%"
"""
cur.execute(sql1)

con.commit()
con.close()
    
```

```python
def to_Int(s) :
    if not s:
        return 0
    return int( "".join(x for x in s if x.isdigit()))
sql5 = "Select recID, recPrepTime, recTotalTime from Recipe"

sql6 = "Update Recipe set recTime = ? where recID = ?"
for t in cur.execute(sql5).fetchall():
    time = to_Int(t[1]) + to_Int(t[2])
    cur.execute(sql6, (time, t[0]))
con.commit()

```
