Importierung der benötigten Module

```python
import sqlite3
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = os.path.realpath('../Data')

con = sqlite3.connect(path+"/RecipeDB.db")
ing_df = pd.read_sql('SELECT ingID FROM Ingredients', con)
al_df = pd.read_sql('SELECT alleID FROM Allergens', con)
tag_df = pd.read_sql('Select tagID FROM Tags', con)
recs_df = pd.read_sql('SELECT recID From Recipe', con)

print( ing_df.shape, "....", al_df.shape, '......', tag_df.shape,'.....', recs_df.shape)
```

```python
tags = [x for x in tag_df['tagID']]
ings = [x for x in ing_df['ingID']]
alls = [x for x in al_df['alleID']]

columns = tags+ings+alls
print(len(columns))
index = [x for x in recs_df['recID']]
print(len(index))
```

```python
data  =np.zeros((len(index), len(columns)), dtype=np.bool_)
data.shape
df= pd.DataFrame(data=data,index = index,columns = columns)
```

```python
df
```

```python


cur = con.cursor()
for ing in cur.execute("SELECT ingID, recID FROM IngInRec ;").fetchall():
    df.loc[ing[1]][ing[0]] = True
    
for ing in cur.execute("SELECT tagID, recID FROM TagsInRec ;").fetchall():
    df.loc[ing[1]][ing[0]] = True
    
for ing in cur.execute("SELECT alleID, recID FROM AlleInRec ;").fetchall():
    df.loc[ing[1]][ing[0]] = True

df.head()
```
